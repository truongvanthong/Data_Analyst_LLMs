```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv(r'D:\CODE_LLM\Data_Analyst_LLMs\data\USA_dataset.csv')

# Create a boxplot for Confirmed cases
plt.figure(figsize=(10,6))
sns.boxplot(y='Confirmed', data=df)
plt.title('Boxplot of Confirmed Cases')
plt.show()

# Create a histogram for Mortality Rate
plt.figure(figsize=(10,6))
sns.histplot(df['Mortality_Rate'].dropna(), bins=20, kde=True)
plt.title('Histogram of Mortality Rate')
plt.xlabel('Mortality Rate')
plt.ylabel('Frequency')
plt.show()
```
